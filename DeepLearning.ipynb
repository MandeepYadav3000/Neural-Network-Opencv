```python
import numpy as np
from numpy import genfromtxt
```

```python
data = genfromtxt('../DATA/DeepLearning/bank_note_data.txt', delimiter=',')
```

```python
len(data)
```

```python
labels = data[:,4]
```

```python
labels
```

```python
features = data[:,0:4]
```

```python
features
```

```python
X = features
y = labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_object = MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaled_X_train = scaler_object.transform(X_train)
```

```python
scaled_X_test = scaler_object.transform(X_test)
```

```python
X_train.max()
```

```python
scaled_X_train.max()
```

```python
scaled_X_train
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
# from keras.models import Sequential
# from keras.layers import Dense
```

```python
from keras.models import Sequential
from keras.layers import Dense
# Creates model
model = Sequential()

# 8 Neurons, expects input of 4 features. 
# Play around with the number of neurons!!
model.add(Dense(4, input_dim=4, activation='relu'))

# Add another Densely Connected layer (every neuron connected to every neuron in the next layer)
model.add(Dense(8, activation='relu'))

# Last layer simple sigmoid function to output 0 or 1 (our label)
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Play around with number of epochs as well gone through data!
model.fit(scaled_X_train,y_train,epochs=50, verbose=2)
```

```python
# model.predict_classes(scaled_X_test)
```

```python
model.metrics_names
```

```python
model.evaluate(x=scaled_X_test,y=y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions = (model.predict(scaled_X_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```
